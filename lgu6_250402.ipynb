# 필터링
- NumPy와 문법 동일

```python
# tips의 평균 구하기
# 평균보다 큰 데이터만 조회
mean_tip = tips['tip'].mean()
# numpy 문법 ==> a[a>12]
tips[tips['tip'] > mean_tip].head(1)
```

```python
# 위에 두줄 식을 한줄로 정리하자면
tips[tips['tip'] > tips['tip'].mean()].head(1)
```

# loc vs iloc
- 코드 비교

##loc

```python
# tips.loc[tips['day'] == 'Sat']
# tips.loc[행, 열]
tips.loc[0:1, ['total_bill', 'tip', 'day']]
```

```python
# iloc
tips.iloc[0:1, [0, 1, 4]]
```

```python
# total_bill 11이하인것만 조회
# tips.loc[:, :]
tips.loc[tips['total_bill'] <= 11, :].head(1)
```

```python
#time이 dinner인 것만 조회, loc 적용
tips.loc[tips['time'] == 'Dinner', :]
```

```python
tips.loc[(tips['time'] == 'Dinner') & (tips['total_bill'] <= 11), :]
# 조건식 2개
# 필터링 문법 : tips.loc[(조건식1) & (조건식2), :]
```

```python
#time이 Dinner이면서 total_bill이 11이하인 것만 조회회
result = tips.loc[tips['time'] == 'Dinner', :]
result.loc[result['total_bill'] <= 11, :]
```

```python
# iris
# 품종이 virginica 이거나 sepal_length >= 5인 값만 가져오고
# 칼럼은 sepal_length, petal_length, species만 가져오세요.
# loc 기반으로

# iris.loc[(조건식1) or (조건식2), ['sepal_length', 'petal_length', 'species']]
# iris.loc[(조건식1) | (조건식2), ['sepal_length', 'petal_length', 'species']]
iris.loc[(iris['species'] == 'virginica') | (iris['sepal_length'] >= 5), ['sepal_length', 'petal_length', 'species']]
```
